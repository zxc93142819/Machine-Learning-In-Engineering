```python
def gradient(x , y , w):
    return (2 * x * (w * x - y)).mean()
```

```python
def gradient_descent(x , y , w , ita , torr = 1.e-4 , max_itr = 5000):
    itr = 1
    dw = 2 * torr

    while abs(dw) > torr and itr < max_itr :
        grad = gradient(x , y , w)
        dw = ita * grad # change in parameter w
        w -= w # updata w
        print(f'Iteration {itr}. w = {w:.2f}.')
        itr += 1
    print("jobs done!")
    print(f'Iteration {itr}. w = {w:.2f}.')
```

```python
gradient_descent(x , y , w = 0.3 , ita = 1.e-5)
```
